### Prepare the dataframe

```python
selected_atlas_type = 'EveType1'
selected_region_id = '140'
selected_measure_type = 'FA_std'

# Use the df_motion as the base to build upon
df_part = df_motion.copy()

# New column on the right to store measure values such as FA_std
new_col_name = selected_measure_type
df_part[new_col_name] = np.nan

# Loop through the columns in the huge spreadsheet, and find the column we want
for col in df_ss.columns:
    # Skip the session column
    if col == 'Session':
        continue
    # Parse the column name 
    atlas_type = col.split(' ')[0].split('-')[0]
    region_id  = col.split(' ')[0].split('-')[1]
    if '-Volume' in col:
        measure_type = 'volume'
        dti_id = None
    elif ('-mean' in col) or ('-std' in col):
        measure_type = "{0}_{1}".format(col.split('-')[-2],col.split('-')[-1])
        # DTI1 or DTI2? (we don't care about DTI_double here)
        if 'DTI1' in col:
            dti_id = 'DTI1'
        elif 'DTI2' in col:
            dti_id = 'DTI2'
        else:
            # make sure that all cases are covered by if
            if not 'DTI_double' in col:
                print('DTI_id specified wrongly.')
                exit()
            else:
                dti_id = 'DTI_double'
    else:
        print('Some measure types are missing.')
        exit()
    
    # Assign every values in this column to the new dataframe
    if (atlas_type == selected_atlas_type) and (region_id == selected_region_id) and (measure_type == selected_measure_type) and (dti_id != 'DTI_double'):
        print(col)
        for session, value in zip(df_ss['Session'], df_ss[col]):
            df_part.loc[(df_part['Session']==session) & (df_part['DTI_ID']==int(dti_id[-1])), new_col_name] = value

# Save the example dataframe
path_project = Path('/home/local/VANDERBILT/gaoc11/Projects/Variance-Aging-Diffusion')
path_to_save = path_project / 'Data' / 'part' / '{0}-{1}-{2}.csv'.format(selected_atlas_type,
                                                                         selected_region_id,
                                                                         selected_measure_type)
# Drop NaN values
df_part.dropna(inplace=True)
df_part.to_csv(path_to_save, index=False)
df_part
```
